```python
%matplotlib inline

import matplotlib
matplotlib.rcParams['figure.figsize'] = [16, 8]
font = {
    'weight': 'normal',
    'size'   : 18}
matplotlib.rc('font', **font)

import matplotlib.pyplot as plt

def plot_mem_line_chart(df, y, save_as):
    ax = df.plot(x='elapsed_secs', y=y)
    ax.set_ylabel('Mb Used')
    plt.savefig(save_as, bbox_inches='tight')
```

```python
import pandas as pd

solution_pandas_df = pd.read_csv(
    'solution_pandas.dat', sep=' ', usecols=[1])

solution_pandas_df['elapsed_secs'] = solution_pandas_df.index / 10
solution_pandas_df['solution_pandas_mb_used'] = solution_pandas_df['python']


solution_pandas_df.head()
```

```python
plot_mem_line_chart(solution_pandas_df, ['solution_pandas_mb_used'], 'solution_pandas.png')
```

```python
solution_loop_df = pd.read_csv(
    'solution_loop.dat', sep=' ', usecols=[1])

solution_loop_df['elapsed_secs'] = solution_loop_df.index / 10
solution_loop_df['solution_loop_mb_used'] = solution_loop_df['python']

solution_loop_df.head()
```

```python
plot_mem_line_chart(solution_loop_df, ['solution_loop_mb_used'], 'solution_loop.png')
```

```python
all_df = pd.concat([solution_pandas_df['solution_pandas_mb_used'], solution_loop_df['solution_loop_mb_used']], axis=1)
all_df['elapsed_secs'] = all_df.index / 10

all_df.head()
```

```python
plot_mem_line_chart(all_df, ['solution_pandas_mb_used', 'solution_loop_mb_used'], 'all.png')
```

```python
solution_pandas_few_cols_df = pd.read_csv(
    'solution_pandas_few_cols.dat', sep=' ', usecols=[1])

solution_pandas_few_cols_df['elapsed_secs'] = solution_pandas_few_cols_df.index / 10
solution_pandas_few_cols_df['solution_pandas_few_cols_mb_used'] = solution_pandas_few_cols_df['python']

solution_pandas_few_cols_df.head()
```

```python
plot_mem_line_chart(solution_pandas_few_cols_df, ['solution_pandas_few_cols_mb_used'], 'solution_pandas_few_cols.png')
```

```python
solution_dask_df = pd.read_csv(
    'solution_dask.dat', sep=' ', usecols=[1])

solution_dask_df['elapsed_secs'] = solution_dask_df.index / 10
solution_dask_df['solution_dask_mb_used'] = solution_dask_df['python']

solution_dask_df.head()
```

```python
solution_pandas_df['solution_pandas_few_cols_mb_used'] = solution_pandas_few_cols_df['solution_pandas_few_cols_mb_used']
solution_pandas_df['solution_dask_mb_used'] = solution_dask_df['solution_dask_mb_used']

plot_mem_line_chart(solution_pandas_df, ['solution_pandas_mb_used', 'solution_pandas_few_cols_mb_used', 'solution_dask_mb_used'], 'solution_pandas_few_cols.png')
```
